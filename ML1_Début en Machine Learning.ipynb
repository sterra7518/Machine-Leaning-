# Ouverutre du jeux de données 

Le but du ML est donc d'entraîner un modèle prédictif à partir d'un jeu de données.

Quand on parle de jeux de données, pensez à une feuille de type Google Spreadsheet ou tableur Excel. Les variables sont les colonnes, et les échantillons sont les lignes.On distingue la variable cible, sujet de la prédiction, des autres variables potentiellement prédictrices.

On part d'un jeu de données à priori “exploitable”. Concept fourre-tout qui suppose que les variables aient une relation entre elles. Le dataset n'est pas un regroupement totalement aléatoire de données qui auraient été collectées on ne sait comment.

Par exemple, prenons un jeu de données comprenant l'âge, la taille et le poids d'une centaine de collégiens. Si on souhaite prédire le poids des enfants en fonction de leur taille et de leur âge, la variable cible sera le poids et les variables prédictrices seront l'âge et la taille. On suppose raisonnablement que la taille et l'âge d'un enfant sont liés à son poids.



# Découvrez les notions de modèle et d’algorithme

On va distinguer :

    le type de modèle : XGBoost, NN, LR : il s'agit là de la méthode issue d'un raisonnement mathématique et traduite en code ;

### Le type de modèle

En Python on importe le modèle en tant que librairie. Exemple avec le modèle de régression linéaire :


```python
from sklearn.ensemble import LinearRegression
```

### L’instanciation du modèle après entraînement

En Python, cela consiste à utiliser la méthode  fit()  :

```python
# Soit des données artificielles
X = [[0, 0], [1, 1]]
y = [0, 1]

# la classe du modèle
from sklearn.linear_model import LinearRegression

# l'instanciation du modèle
mdl = LinearRegression()
mdl = mdl.fit(X, y)
```

```python
mdl
```

Dans le code ci-dessus,  clf  est le modèle instancié, qui est entraîné sur les données  X  et  y  .

Par souci de simplicité on parlera de modèle dans les 2 cas, classe et instance, mais il est bon de souligner la différenc

```python
# initialisation
x = 1
precision = 0.001
# boucle itérative
while (abs(x**2 - 2) > precision) :
    x = (x + 2/x)/2

print(x)
```

Remarquez maintenant la quantité [Math Processing Error]
  que nous avons calculée à chaque itération : la différence en valeur absolue entre la valeur candidate au carré [Math Processing Error]
  et 2. C'est l'erreur d'estimation entre la valeur estimée et la valeur dite de vérité (groundtruth, en anglais). On peut réécrire l'algorithme en mettant en avant cette erreur d'estimation 

```python
# initialisation
x = 1
erreur = 1
precision = 0.001


# boucle itérative
while erreur > precision:
    # mise à jour de l'estimation
    x = (x + 2/x)/2
    # mise à jour de l'erreur avec la nouvelle valeur de x
    erreur = abs(x**2 - 2)
```
